```python
import psycopg2
import numpy as np
import psycopg2.extras as extras
import pandas as pd
import os
from sqlalchemy import create_engine


engine = create_engine(os.environ['COCKROACH_URL'])
engine.connect()
```

## Techies

```python
techies = pd.read_csv('../data/techies.csv')
techies.head(20)
```

```python
techies.to_sql("techies", con=engine, if_exists="fail", index=False)
```
